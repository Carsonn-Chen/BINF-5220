```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/BINF 5220/data/hba1c_bloodGlucose.dat", sep='\s+', skiprows=30)
print(df)
print(df.columns)
print(len(df))
```

```python
def sub_divide(x):
  x -= 2
  return x/2, x/4

for loc in range(0, len(df['hba1c']), 10):
  # print(loc)
  print(sub_divide(df['hba1c'][loc]))
```

```python
fig = plt.figure()
ax = fig.subplots()


ax.scatter(df['rbg'], df['hba1c'], label="hba1c data scatter")
ax.set_xlabel('rbg (mmol/l)')
ax.set_ylabel('hba1c (DCCT percentages)')
ax.legend()
plt.show()
# fig.savefig('fig.png', bbox_inches='tight')
```
